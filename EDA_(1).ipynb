**Visualizations**

Visualize key findings using appropriate plots such as line plots, bar charts, heatmaps, etc.

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10,6))
plt.scatter(x = df["Product Category"] , y = df["Total Amount"])
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# Group data by product category and date, summing up sales
grouped_data = df.groupby(['Product Category', 'Date']).sum().reset_index()

# Plotting
plt.figure(figsize=(12, 6))

# Iterate over each product category
for category, data in grouped_data.groupby('Product Category'):
    plt.plot(df['Date'], df['Product Category'], label=category)

plt.title('Sales by Product Category Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Heatmap: Sales Performance by Product Category and Region
sales_pivot = df.pivot_table(index='Product Category', columns='Date', values='Total Amount', aggfunc=sum)
plt.figure(figsize=(10, 6))
sns.heatmap(sales_pivot, cmap='YlGnBu', annot=True, fmt='.0f')
plt.title('Sales Performance by Product Category and Date ')
plt.xlabel('Date')
plt.ylabel('Product Category')
plt.show()
```

```python
# Heatmap: Sales Performance by Product Category and Region
sales_pivot = df.pivot_table(index='Product Category', columns='Quantity', values='Total Amount', aggfunc=sum)
plt.figure(figsize=(10, 6))
sns.heatmap(sales_pivot, cmap='YlGnBu', annot=True, fmt='.0f')
plt.title('Sales Performance by Product Category and Quantity')
plt.xlabel('Quantity')
plt.ylabel('Product Category')
plt.show()
```

```python
# Bar Chart: Sales by Product Category
plt.figure(figsize=(10, 6))
sns.barplot(x='Product Category', y='Total Amount', data=df, estimator=sum)
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

**MC donald's Nutrition Analysis**

The dataset under examination provides a comprehensive nutritional breakdown of each menu item available on the McDonald's menu in the United States. Encompassing a diverse array of offerings ranging from breakfast items to beef burgers, chicken and fish sandwiches, fries, salads, beverages such as soda, coffee, and tea, as well as indulgent treats like milkshakes and desserts, this dataset furnishes a comprehensive view of the nutritional landscape of McDonald's offerings. Notably, the menu items and associated nutritional information were systematically extracted from the official McDonald's website, ensuring the accuracy and reliability of the data utilized in this analysis.

```python
data = pd.read_csv("/content/menu.csv")
data.head(5)
```

```python
data.shape
```

```python
plt.figure(figsize=(10, 4), dpi=100)
menu_category = data.Category.value_counts()
menu_category.plot.bar(color = ['red','pink','pink','pink','pink','pink','pink','pink','pink'])
plt.title("Number of Menu Items for each Food Category")
plt.ylabel("Count")
plt.xlabel("Menu Category")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 4), dpi=100)

correlation = data['Calories'].corr(data['Calories from Fat'])
plt.scatter(data.Calories, data['Calories from Fat'], color='blue')
plt.text(500,450,'r = {}'.format(round(correlation,2)))
plt.xlabel("Calories")
plt.ylabel("Calories from Fat")
plt.title("Relationship between Calories and Calories from Fat")
plt.show()
```

```python
data.groupby('Category')['Trans Fat'].mean()
```

```python
data['saturated_cholesterol'] = data['Saturated Fat']/data['Cholesterol']*100
```

```python
saturated_cholesterol = data.groupby('Category')['saturated_cholesterol'].mean().dropna().nlargest(5)
```

```python
plt.figure(figsize=(10, 4), dpi=100)
saturated_cholesterol.sort_values(ascending=False).plot.bar(color = 'orange')
plt.title("Top 5 food categories with the highest % Saturated Fat")
plt.ylabel("Percentage")
plt.xlabel("Menu Category")
plt.xticks(rotation=45)
plt.show();
```

```python
plt.figure(figsize=(10, 4), dpi=100)
plt.scatter(data['Total Fat (% Daily Value)'], data['Carbohydrates (% Daily Value)'], color='purple')
correlation = data['Total Fat (% Daily Value)'].corr(data['Carbohydrates (% Daily Value)'])
plt.text(125,20,'r = {}'.format(round(correlation,2)))
plt.xlabel("% Total Fat")
plt.ylabel("% Total Carbohydrate")
plt.title("Relationship between % Total Fat and % Total Carbohydrate")
plt.show()
```

```python
iron = data.groupby('Category')['Iron (% Daily Value)'].mean()
calcium = data.groupby('Category')['Calcium (% Daily Value)'].mean()
```

```python
plt.figure(figsize=(10, 4), dpi=100)

category = iron.index
Iron = iron.values
Calcium = calcium.values

X_axis = np.arange(len(category))

plt.bar(X_axis - 0.2, Iron, 0.4, label = 'Iron', color = 'pink')
plt.bar(X_axis + 0.2, Calcium, 0.4, label = 'Calcium', color = 'black')

plt.xticks(X_axis, category)
plt.xlabel("Menu Categories")
plt.title("Distribution of Iron and Calcium in McDonald's Foods")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
